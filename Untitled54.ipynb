```python
import numpy as np
import matplotlib.pyplot as plt

def z(t):
    return (np.heaviside(t+2, 1) - np.heaviside(t+1, 1)) + \
           2*(np.heaviside(t-1, 1) - np.heaviside(t-2, 1)) - \
           (np.heaviside(t-2, 1) - np.heaviside(t-3, 1))
def r(t):
    return np.where(t >= 0, t, 0)
def w(t):
    return z(t) * r(2*(t+12)-6)
t = np.linspace(-15, 15, 1000)

y = w(t)

plt.plot(t, y)
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.title('Gráfica de la señal w(t)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#A
def segment_signal(x, M, S):
    step = M - S
    segments = [x[i:i + M] for i in range(0, len(x) - M + 1, step)]
    return np.array(segments)

#B
def compute_dft_segments(segments, w):
    N = len(w)
    periodograms = []
    for seg in segments:
        windowed_seg = seg * w
        dft = np.fft.fft(windowed_seg, N)
        periodogram = (np.abs(dft)**2) / N
        periodograms.append(periodogram[:N // 2])
    return np.array(periodograms)

#C
def welch_periodogram(x, fs, M, S, window_type='hann'):
    if window_type == 'hann':
        w = np.hanning(M)
    elif window_type == 'rectangular':
        w = np.ones(M)
    else:
        raise ValueError("Ventana no soportada.")

    segments = segment_signal(x, M, S)

    periodograms = compute_dft_segments(segments, w)
#D Se usa la función np.mean para calcular el promedio de todos los periodogramas
    average_periodogram = np.mean(periodograms, axis=0)

    freqs = np.linspace(0, fs / 2, len(average_periodogram))
    return freqs, average_periodogram

if __name__ == "__main__":
    a = 5
    k = 2 * (a + 1)
    fs = 200
    T = 10
    t = np.linspace(0, T, T * fs, endpoint=False)
    M = 256
    S = 128

    #señal
    x_t = 4 * np.sin(80 * np.pi * t + np.pi / 4) + k * np.cos(40 * np.pi * t) + 5

    #periodograma de Welch
    freqs, average_periodogram = welch_periodogram(x_t, fs, M, S)

    plt.figure(figsize=(10, 6))
    plt.plot(freqs, 10 * np.log10(average_periodogram))
    plt.title("Periodograma de Welch")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Densidad espectral de potencia (dB)")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Validación de la rutina con los parámetros indicados
def validate_welch_routine():
    # Parámetros de la señal
    a = 5
    k = 2 * (a + 1)  # Cálculo de k
    fs = 200  # Frecuencia de muestreo (Hz)
    T = 10  # Duración de la señal en segundos
    t = np.linspace(0, T, T * fs, endpoint=False)  # Vector de tiempo

    # Definición de la señal x(t)
    x_t = 4 * np.sin(80 * np.pi * t + np.pi / 4) + k * np.cos(40 * np.pi * t) + 5

    # Parámetros para la rutina de Welch
    M = 40  # Muestras por segmento
    S = 25  # Muestras de solapamiento
    w = np.hanning(M)  # Ventana Hann

    # Segmentación de la señal
    step = M - S  # Avance por segmento
    segments = [x_t[i:i + M] for i in range(0, len(x_t) - M + 1, step)]
    segments = np.array(segments)

    # Aplicación de la ventana y cálculo de la DFT
    N = len(w)
    periodograms = []
    for seg in segments:
        windowed_seg = seg * w  # Multiplicación por la ventana
        dft = np.fft.fft(windowed_seg, N)  # Transformada de Fourier
        periodogram = (np.abs(dft)**2) / N  # Periodograma
        periodograms.append(periodogram[:N // 2])  # Mitad positiva del espectro
    periodograms = np.array(periodograms)

    # Promedio de los periodogramas
    average_periodogram = np.mean(periodograms, axis=0)

    # Frecuencias asociadas
    freqs = np.linspace(0, fs / 2, len(average_periodogram))

    # Graficar el resultado
    plt.figure(figsize=(10, 6))
    plt.plot(freqs, 10 * np.log10(average_periodogram))
    plt.title("Validación: Periodograma de Welch (Ventana Hann)")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Densidad espectral de potencia (dB)")
    plt.grid(True)
    plt.show()

# Ejecutar la validación
validate_welch_routine()
```

El gráfico muestra los picos en las frecuencias dominantes de la señal en 40Hz y 20Hz
Como el periodograma de Welch muestra cómo la energía de la señal se distribuye en diferentes frecuencias. Su uso de ventanas y promedios asegura una estimación más estable y precisa, especialmente útil para señales no estacionarias o contaminadas con ruido. observamos los picos entre 20HZ y 40HZ correctamente confirmamos que la rutina funciona correctamente

7)utilice 49 Hz en lugar de 55 Hz porque la frecuencia de corte del filtro debe estar estrictamente dentro del rango válido, que va de fs/2 (en este caso de 50 Hz porque la frecuencia de muestreo es fs=100 Hz) El límite superior
fs/2=50 Hz no es permitido por la función firwin ya que esta frecuencia corresponde a Nyquist y los filtros diseñados con un número impar de coeficientes no pueden tener respuesta en esa frecuencia.por lo tanto elegi 49HZ

```python
from scipy.signal import firwin, freqz

fs = 100
a = 10
b = 49

numtaps_high = 101
cutoff_high = a
highpass = firwin(numtaps=numtaps_high, cutoff=cutoff_high, pass_zero=False, fs=fs, window='hamming')

numtaps_low = 101
cutoff_low = b
lowpass = firwin(numtaps=numtaps_low, cutoff=cutoff_low, pass_zero=True, fs=fs, window='hamming')

frequencies_high, response_high = freqz(highpass, worN=8000, fs=fs)
frequencies_low, response_low = freqz(lowpass, worN=8000, fs=fs)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
plt.subplots_adjust(hspace=0.4)

#Magnitu altas
axs[0, 0].plot(frequencies_high, 20 * np.log10(np.abs(response_high)), label="Pasa-altas")
axs[0, 0].set_title("Filtro pasa-altas: Magnitud")
axs[0, 0].set_xlabel("Frecuencia (Hz)")
axs[0, 0].set_ylabel("Magnitud (dB)")
axs[0, 0].grid(True)
axs[0, 0].legend()

#Fase altas
axs[1, 0].plot(frequencies_high, np.unwrap(np.angle(response_high)), label="Pasa-altas")
axs[1, 0].set_title("Filtro pasa-altas: Fase")
axs[1, 0].set_xlabel("Frecuencia (Hz)")
axs[1, 0].set_ylabel("Fase (rad)")
axs[1, 0].grid(True)
axs[1, 0].legend()

#Magnitud bajas
axs[0, 1].plot(frequencies_low, 20 * np.log10(np.abs(response_low)), label="Pasa-bajas")
axs[0, 1].set_title("Filtro pasa-bajas: Magnitud")
axs[0, 1].set_xlabel("Frecuencia (Hz)")
axs[0, 1].set_ylabel("Magnitud (dB)")
axs[0, 1].grid(True)
axs[0, 1].legend()

#Fase bajas
axs[1, 1].plot(frequencies_low, np.unwrap(np.angle(response_low)), label="Pasa-bajas")
axs[1, 1].set_title("Filtro pasa-bajas: Fase")
axs[1, 1].set_xlabel("Frecuencia (Hz)")
axs[1, 1].set_ylabel("Fase (rad)")
axs[1, 1].grid(True)
axs[1, 1].legend()

plt.show()
```
